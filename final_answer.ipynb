```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

x_train=pd.read_csv('/content/training_set_features.csv')
y_train=pd.read_csv('/content/training_set_labels.csv')
x_test=pd.read_csv('/content/test_set_features.csv')
x_test.head()
```

```python

married=pd.get_dummies(x_test['marital_status'],drop_first=True).astype('int')
x_test=pd.concat([x_test,married],axis=1)
x_test.drop(['marital_status'],axis=1,inplace=False)


married=pd.get_dummies(x_train['marital_status'],drop_first=True).astype('int')
x_train=pd.concat([x_train,married],axis=1)
x_train.drop(['marital_status'],axis=1,inplace=False)

x_train.head()
```

```python
x_train.employment_status.value_counts()
x_train.employment_status.fillna(x_train.employment_status.mode()[0],inplace=True)
x_test.employment_status.fillna(x_test.employment_status.mode()[0],inplace=True)
```

```python
dict={'Employed':2,'Not in Labor Force':1,'Unemployed':0}
x_train['Employed']=x_train['employment_status'].map(dict).astype('int')
x_test['Employed']=x_test['employment_status'].map(dict).astype('int')
```

```python

married=pd.get_dummies(x_test['sex'],drop_first=True).astype('int')
x_test=pd.concat([x_test,married],axis=1)
x_test.drop(['sex'],axis=1,inplace=False)


married=pd.get_dummies(x_train['sex'],drop_first=True).astype('int')
x_train=pd.concat([x_train,married],axis=1)
x_train.drop(['sex'],axis=1,inplace=False)

```

```python
for category in x_train.race.unique():
  x_train[category]=np.where(x_train['race']==category,1,0)
  x_test[category]=np.where(x_test['race']==category,1,0)
x_train.head()
```

```python

married=pd.get_dummies(x_test['rent_or_own'],drop_first=True).astype('int')
x_test=pd.concat([x_test,married],axis=1)
x_test.drop(['rent_or_own'],axis=1,inplace=False)


married=pd.get_dummies(x_train['rent_or_own'],drop_first=True).astype('int')
x_train=pd.concat([x_train,married],axis=1)
x_train.drop(['rent_or_own'],axis=1,inplace=False)

```

```python
x_test.age_group.value_counts()

for category in x_train.age_group.unique().astype('str'):
  x_train[category]=np.where(x_train['age_group']==category,1,0)
  x_test[category]=np.where(x_test['age_group']==category,1,0)
x_train.head()
```

```python

x_train.education.unique().astype('str')
for category in x_train.education.unique():
  x_train[category]=np.where(x_train['education']==category,1,0)
  x_test[category]=np.where(x_test['education']==category,1,0)
x_train.head()
```

```python

for category in x_train.income_poverty.unique().astype('str'):
  x_train[category]=np.where(x_train['income_poverty']==category,1,0)
  x_test[category]=np.where(x_test['income_poverty']==category,1,0)
x_train.head()
```

```python
x_train.hhs_geo_region.value_counts()

for category in x_train.hhs_geo_region.unique().astype('str'):
  x_train[category]=np.where(x_train['hhs_geo_region']==category,1,0)
  x_test[category]=np.where(x_test['hhs_geo_region']==category,1,0)
x_train.head()
```

```python
for category in x_train.census_msa.unique().astype('str'):
  x_train[category]=np.where(x_train['census_msa']==category,1,0)
  x_test[category]=np.where(x_test['census_msa']==category,1,0)
x_train.head()
```

```python
x_train.drop(['health_insurance','employment_industry','employment_occupation','census_msa',np.NaN,'hhs_geo_region','income_poverty','education','age_group','rent_or_own','race','employment_status','sex','marital_status'],axis=1,inplace=True)
x_test.drop(['health_insurance','employment_industry','employment_occupation','census_msa',np.NaN,'hhs_geo_region','income_poverty','education','age_group','rent_or_own','race','employment_status','sex','marital_status'],axis=1,inplace=True)
x_test.drop(['respondent_id'],inplace=True,axis=1)
x_train.drop(['respondent_id'],inplace=True,axis=1)

x_train.head()
```
